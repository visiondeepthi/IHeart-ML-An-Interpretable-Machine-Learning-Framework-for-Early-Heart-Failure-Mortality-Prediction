```python

print("\n10. STATISTICAL COMPARISON OF MODELS")
print("-" * 70)


all_results = {}
all_results.update(ml_results)
all_results.update(dl_results)


y_true = y_test.to_numpy()



def mcnemar_test(pred_a, pred_b, y_true):
    """
    McNemar test for two classifiers evaluated on the same samples.
    Returns (chi2_stat, p_value).
    """
    pred_a = np.asarray(pred_a)
    pred_b = np.asarray(pred_b)
    y_true = np.asarray(y_true)

    correct_a = (pred_a == y_true)
    correct_b = (pred_b == y_true)

    n01 = np.sum((correct_a == 1) & (correct_b == 0))  # A correct, B wrong
    n10 = np.sum((correct_a == 0) & (correct_b == 1))  # A wrong, B correct
    n11 = np.sum((correct_a == 1) & (correct_b == 1))  # both correct
    n00 = np.sum((correct_a == 0) & (correct_b == 0))  # both wrong

    table = np.array([[n11, n10],
                      [n01, n00]])

    result = mcnemar(table, exact=False, correction=True)
    return result.statistic, result.pvalue, (n11, n10, n01, n00)

print("\n=== McNemar's Test (Accuracy Comparison) ===")
baseline_name = "Random Forest"   # we compare everything against RF

if baseline_name not in all_results:
    print(f"Baseline model '{baseline_name}' not found!")
else:
    base_pred = all_results[baseline_name]["predictions"]

    for name, res in all_results.items():
        if name == baseline_name:
            continue
        chi2, pval, counts = mcnemar_test(base_pred, res["predictions"], y_true)
        n11, n10, n01, n00 = counts
        print(f"{baseline_name} vs {name}: "
              f"chi2 = {chi2:.3f}, p = {pval:.4f} "
              f"(n11={n11}, n10={n10}, n01={n01}, n00={n00})")



def compute_midrank(x):
    """
    Computes midranks for a 1D array.
    """
    x = np.asarray(x)
    sorted_idx = np.argsort(x)
    sorted_x = x[sorted_idx]
    n = len(x)
    ranks = np.zeros(n, dtype=float)
    i = 0
    while i < n:
        j = i
        while j < n and sorted_x[j] == sorted_x[i]:
            j += 1
        mid = 0.5 * (i + j - 1) + 1  # +1 for 1-based rank
        ranks[i:j] = mid
        i = j
    out = np.empty(n, dtype=float)
    out[sorted_idx] = ranks
    return out

def fast_delong(predictions, labels):
    """
    Fast DeLong implementation for AUC variance.
    predictions: 2D array (n_classifiers, n_samples)
    labels: 1D array (0 or 1)
    Returns (auc, auc_covariance_matrix).
    """
    labels = np.asarray(labels)
    predictions = np.asarray(predictions)
    assert np.array_equal(np.unique(labels), [0, 1]), "Labels must be 0/1."

    # Separate positive and negative samples
    pos_idx = labels == 1
    neg_idx = labels == 0

    pos_preds = predictions[:, pos_idx]
    neg_preds = predictions[:, neg_idx]

    m = pos_preds.shape[1]
    n = neg_preds.shape[1]

    # Rank all samples together per classifier
    all_preds = np.hstack((pos_preds, neg_preds))
    v = np.zeros((predictions.shape[0], m))
    w = np.zeros((predictions.shape[0], n))

    for r in range(predictions.shape[0]):
        ranks = compute_midrank(all_preds[r])
        v[r] = ranks[:m]
        w[r] = ranks[m:]

    # According to DeLong paper
    v = (v - (m + n + 1) / 2.0) / n
    w = (w - (m + n + 1) / 2.0) / m

    aucs = v.mean(axis=1)
    sv = np.cov(v)
    sw = np.cov(w)
    s = sv / m + sw / n
    return aucs, s

def delong_roc_test(labels, preds1, preds2):
    """
    DeLong test for two correlated ROC AUCs.
    Returns (z_stat, p_value, auc1, auc2).
    """
    preds = np.vstack((preds1, preds2))
    aucs, cov = fast_delong(preds, labels)
    diff = aucs[0] - aucs[1]
    var = cov[0, 0] + cov[1, 1] - 2 * cov[0, 1]
    if var <= 0:
        # numerical safety
        z = 0.0
        p = 1.0
    else:
        z = diff / np.sqrt(var)
        # two-sided p-value
        from scipy.stats import norm
        p = 2 * (1 - norm.cdf(abs(z)))
    return z, p, aucs[0], aucs[1]

print("\n=== DeLong's Test (AUC Comparison) ===")

try:
    from scipy.stats import norm  # ensure available
except ImportError:
    print("scipy is required for DeLong's test. Install via 'pip install scipy'.")
else:
    if baseline_name not in all_results:
        print(f"Baseline model '{baseline_name}' not found for DeLong test.")
    else:
        base_prob = all_results[baseline_name]["probabilities"]

        for name, res in all_results.items():
            if name == baseline_name:
                continue
            z_stat, pval, auc1, auc2 = delong_roc_test(
                y_true,
                base_prob,
                res["probabilities"]
            )
            print(f"{baseline_name} (AUC={auc1:.3f}) vs {name} (AUC={auc2:.3f}): "
                  f"z = {z_stat:.3f}, p = {pval:.4f}")
```
